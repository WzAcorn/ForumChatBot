## googletrans을 사용하여 영어로 통일시키는 코드.

```python
from googletrans import Translator, LANGUAGES
import langdetect

# Initialize the translator
translator = Translator()

def translate_text(text):
    try:
        # Detect the language of the text
        detected_language = langdetect.detect(text)
        # If the text is not in English, translate it
        if detected_language != 'en':
            translated_text = translator.translate(text, dest='en').text
            return translated_text
        else:
            # Return the original text if it's already in English
            return text
    except Exception as e:
        # In case of any error, return the original text
        return text

# Apply the translation function to a small subset of the data for testing
df['translated_raw'] = df['raw'].apply(translate_text)

df[['raw', 'translated_raw']]
df.sort_values(by='post_number').sort_values(by='topic_id').head(30)
```

## keywords이 있는 내용들은 삭제.
ex: 이미지나 dbdump들


```python
keywords = ["img", "upload", "png", "jpg", "jpeg", "gif", "bmp", "svg", "file", "attach",
            "Replace this first paragraph",
            "Edit the first post in this topic"]
# Join keywords into a single string for regex search, separate them with '|'
keywords_pattern = '|'.join(keywords)

# Invert the boolean mask to keep rows NOT containing any of the keywords
mask = ~df['raw'].str.contains(keywords_pattern, na=False, case=False, regex=True)
df = df.loc[mask]
# Sort by 'post_number' and then 'topic_id'
sorted_data = df.sort_values(by=['topic_id', 'post_number'])



```

```python
filtered_df = df[df.duplicated('topic_id', keep=False)]
sorted_data = filtered_df.sort_values(by=['topic_id', 'post_number'])


print(sorted_data.drop('raw', axis=1))
sorted_data = sorted_data.drop('raw', axis=1)
sorted_data.to_csv("forum_dataset_reduced.csv")
```

## 데이터 그룹핑

```python
import pandas as pd

# 파일 불러오기
file_path = 'cleaned_forum_dataset.csv'
# 파일을 UTF-8 인코딩으로 불러오기
df_utf8 = pd.read_csv(file_path, encoding='utf-8')

# 데이터의 처음 몇 줄을 표시하여 확인
df_utf8.head()

# 불필요한 Unnamed 칼럼 제거
df_utf8_clean = df_utf8.drop(df_utf8.filter(regex='Unnamed').columns, axis=1)

# translated_raw 칼럼에서 문자열이 아닌 값들을 빈 문자열로 대체
df_utf8_clean['translated_raw'] = df_utf8_clean['translated_raw'].fillna('')

# topic_id 별로 translated_raw 데이터 합치기
df_utf8_grouped = df_utf8_clean.groupby('topic_id')['translated_raw'].apply(' '.join).reset_index()

# 결과 확인
df_utf8_grouped.head()

# 'topic_id'가 숫자 형태인지 확인 (NaN이나 이상한 값이 아닌 경우)
df_utf8_clean = df_utf8_clean[pd.to_numeric(df_utf8_clean['topic_id'], errors='coerce').notna()]

# 다시 topic_id 별로 translated_raw 데이터 합치기
df_grouped_clean = df_utf8_clean.groupby('topic_id')['translated_raw'].apply(' '.join).reset_index()

# 결과 확인
df_grouped_clean.head()

df_grouped_clean.to_csv('test_cleaned_forum_dataset.csv', encoding='utf-8')

```
